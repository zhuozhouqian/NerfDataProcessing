```python
from scipy.spatial.transform import Rotation
```

```python
shutil.copyfile("2022-12-21-15-57-16-car/rgb/1671609436.769321.png","2022-12-21-15-57-16-car/rgb.png")
```

```python
rgb_path="2022-12-21-15-57-16-car/rgb/"
depth_path="2022-12-21-15-57-16-car/depth/"
```

```python
KeyFrameTrajs=np.loadtxt("2022-12-21-15-57-16-car/KeyFrameTrajectory.txt")
```

```python
KeyFrameTrajs=np.array(KeyFrameTrajs)
```

```python
KeyFrameTrajs
```

```python
KeyFrameTrajs_Rotation=KeyFrameTrajs[:,4:8]
```

```python
Timestamp=KeyFrameTrajs[:,0]
```

```python
nums=np.size(KeyFrameTrajs_Rotation,axis=0)
```

```python
KeyFrameTrajs_Matrix=np.zeros([nums,4,4])
```

```python
KeyFrameTrajs_Matrix=np.zeros([nums,4,4])
```

```python
for i in range(0,nums):
    Rm=Rotation.from_quat(KeyFrameTrajs_Rotation[i,:])
    KeyFrameTrajs_Matrix[i,0:3,0:3]=Rm.as_matrix()
    KeyFrameTrajs_Matrix[i,0:3,3]=KeyFrameTrajs[i,1:4]
    KeyFrameTrajs_Matrix[i,3,:]=[0,0,0,1]
```

```python
np.size(KeyFrameTrajs_Matrix,axis=0)
```

```python
os.mkdir("output")
os.mkdir("output/color")
os.mkdir("output/depth")
```

```python
os.mkdir("output/intrinsic")
```

```python
os.path.join(rgb_path,str(Timestamp[0])+".png")
```

```python
'%.6f' % Timestamp[0]
```

```python
os.mkdir("output/pose")
```

```python
for i in range(0,nums):
    shutil.copyfile(os.path.join(rgb_path,'%.6f' % Timestamp[i]+".png"),os.path.join("output/color",str(i)+".png"))
    shutil.copyfile(os.path.join(depth_path,'%.6f' % Timestamp[i]+".png"),os.path.join("output/depth",str(i)+".png"))
    np.savetxt(os.path.join("output/pose",str(i)+".txt"),KeyFrameTrajs_Matrix[i])
```
